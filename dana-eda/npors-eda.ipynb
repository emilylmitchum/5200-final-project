### Possible Visualization 1

A pie chart that shows distribution of responses for different issues, then when user clicks on a slice we can see how many of those respondents were men vs. women vs. etc.

```python
df = npors_2025_clean

#DATA PREPARATION
df_agg = df.groupby(['ECON1MOD', 'GENDER']).size().reset_index(name='Total_Count')

display(df_agg)
```

```python
import altair as alt

#INTERACTIVE SELECTION
click = alt.selection_point(fields=['ECON1MOD'])

#PIE CHART
pie = alt.Chart(df_agg).mark_arc().encode(
    theta=alt.Theta('sum(Total_Count):Q', stack=True),
    color=alt.Color('ECON1MOD:N', legend=alt.Legend(title="ECON1MOD Responses")),
    opacity=alt.condition(click, alt.value(1.0), alt.value(0.3)),
    tooltip=['ECON1MOD:N', alt.Tooltip('sum(Total_Count):Q', title='Total Count')]
).add_params(
    click
).properties(
    width=400,
    height=400,
    title='Distribution of Responses (ECON1MOD)'
)

#BAR CHART
bar = alt.Chart(df_agg).mark_bar().encode(
    x=alt.X('GENDER:N', title='Gender', sort='-y', axis=alt.Axis(labelAngle=0)),
    y=alt.Y('sum(Total_Count):Q', title='Number of People'),
    color=alt.Color('GENDER:N', legend=None),
    tooltip=['GENDER:N', alt.Tooltip('sum(Total_Count):Q', title='Total Count')]
).transform_filter(
    click #links the two views together
).properties(
    width=350,
    height=350,
    title='Gender of Selected Group'
)


#ARRANGE AND DISPLAY
interactive_chart = pie | bar
interactive_chart
```

### Possible Visualization 1.2

Same as above but we can see multiple issues in this linked view

```python
#DATA PREPARATION
issues_to_toggle = ['ECON1MOD', 'CRIMESAFE', 'MOREGUNIMPACT', 'FIN_SIT', 'RELIMP']
melted_df = df.melt(
    id_vars=['GENDER'], 
    value_vars=issues_to_toggle, 
    var_name='Issue', 
    value_name='Response'
)


df_agg = melted_df.groupby(['Issue', 'Response', 'GENDER']).size().reset_index(name='Total_Count')
display(df_agg)
```

```python
#DROPDOWN SELECTION
issue_dropdown = alt.binding_select(options=issues_to_toggle, name='Select Issue: ')
issue_select = alt.selection_point(
    fields=['Issue'], 
    bind=issue_dropdown, 
    value=[{'Issue': 'ECON1MOD'}]
)

#INTERACTIVE SELECTION
click = alt.selection_point(fields=['Response'])

#PIE CHART
pie = alt.Chart(df_agg).mark_arc().encode(
    theta=alt.Theta('sum(Total_Count):Q', stack=True),
    color=alt.Color('Response:N', legend=alt.Legend(title="Response Options", orient="bottom", columns=1)),
    opacity=alt.condition(click, alt.value(1.0), alt.value(0.3)),
    tooltip=['Response:N', alt.Tooltip('sum(Total_Count):Q', title='Total Count')]
).add_params(
    issue_select,
    click
).transform_filter(
    issue_select
).properties(
    width=350,
    height=350,
    title='Distribution of Responses'
)

#BAR CHART
bar = alt.Chart(df_agg).mark_bar().encode(
    x=alt.X('GENDER:N', title='Gender', sort='-y', axis=alt.Axis(labelAngle=0)),
    # CHANGE: Map Y directly to the sum of our Total_Count column
    y=alt.Y('sum(Total_Count):Q', title='Number of People'),
    color=alt.Color('GENDER:N', legend=None),
    tooltip=['GENDER:N', alt.Tooltip('sum(Total_Count):Q', title='Total Count')]
).transform_filter(
    issue_select
).transform_filter(
    click
).properties(
    width=350,
    height=350,
    title='Gender of Selected Group'
)

#ARRANGE AND DISPLAY
interactive_chart = pie | bar
interactive_chart
```
