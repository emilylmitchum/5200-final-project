### Possible Visualization 1.2

Same as above but we can see multiple issues in this linked view

ECON1MOD = "How would you rate the economic conditions of your community today?
CRIMESAFE = "How would you describe the area where you live, in terms of crime?"
MOREGUNIMPACT = "If more Americans owned guns, do you think there would be ..."
FIN_SIT = "How would you describe your household's financial situation?"
RELIMP = "How important is religion in your life?"
GOVPROTCT = "Which statement comes closer to your own view, even if neither is exactly right?"


```python
#DATA PREPARATION
issues_to_toggle = ['ECON1MOD', 
                    'ECON1BMOD',
                    'CRIMESAFE', 
                    'MOREGUNIMPACT', 
                    #'FIN_SIT', #delete because not an issue but could be interesting
                    #'RELIMP', #delte for same reason above
                    'GOVPROTCT']
melted_df = df.melt(
    id_vars=['GENDER'], 
    value_vars=issues_to_toggle, 
    var_name='Issue', 
    value_name='Response'
)


df_agg = melted_df.groupby(['Issue', 'Response', 'GENDER']).size().reset_index(name='Total_Count')
display(df_agg)
```

```python
#DROP ROWS with "refused/blank" or "in some other way" in gender for easier viewing

df_agg_clean = df_agg[
    (df_agg['GENDER'] != "In some other way") & 
    (df_agg['GENDER'] != "Refused/Web blank") & 
    (df_agg['Response'] != "Refused/Web blank")
].copy()


#ADD COLUMN WITH READABLE ISSUE
issue_map = {
    "ECON1MOD": "Economic conditions today",
    "ECON1BMOD": "Future economic conditions",
    "CRIMESAFE": "Safety from crime", 
    "MOREGUNIMPACT": "Impact of more Americans owning guns", 
    "GOVPROTCT": "Responsibility to protect people from themselves"
}

df_agg_clean.loc[:, 'Readable_Issue'] = df_agg_clean['Issue'].map(issue_map)

#UPDATE RESPONSE OF FINAL ISSUE FOR READABILITY
# Create a dictionary of { "Old Text": "New Text" }
text_fixes = {
    "It's not the government's job to protect people from themselves": "Not government",
    "Sometimes laws to protect people from themselves are necessary":"Government"
    # You can add more here if you need to!
    # "Another really long survey response": "Short version" 
}

# Apply the replacement directly to the Response column
df_agg_clean['Response'] = df_agg_clean['Response'].replace(text_fixes)

display(df_agg_clean)
```

```python
readable_issues_to_toggle = [
    "Economic conditions today",
    "Future economic conditions",
    "Safety from crime", 
    "Impact of more Americans owning guns", 
    "Responsibility to protect people from themselves"
]
```

```python
#COLORS
my_custom_palette = ["#ef476f", "#ffd166", "#f78c6b", "#06d6a0", "#118ab2"]


#DROPDOWN SELECTION
issue_dropdown = alt.binding_select(options=readable_issues_to_toggle, name='Select Issue: ')
issue_select = alt.selection_point(
    fields=['Readable_Issue'], 
    bind=issue_dropdown, 
    value=[{'Readable_Issue': 'Economic conditions today'}]
)

#INTERACTIVE SELECTION
click = alt.selection_point(fields=['Response'])



#PIE CHART
pie = alt.Chart(df_agg_clean).mark_arc().encode(
    theta=alt.Theta('sum(Total_Count):Q', stack=True),
    color=alt.Color(
                    'Response:N',
                    legend=alt.Legend(title="Response Options", orient="right", columns=1),
                    scale=alt.Scale(range=my_custom_palette)
                    ),
    opacity=alt.condition(click, alt.value(1.0), alt.value(0.3)),
    tooltip=['Response:N', alt.Tooltip('sum(Total_Count):Q', title='Total Count')]
).add_params(
    issue_select,
    click
).transform_filter(
    issue_select
).properties(
    width=300,
    height=300,
    title='Distribution of Responses to Issue'
)

#BAR CHART
bar = alt.Chart(df_agg_clean).mark_bar().encode(
    x=alt.X('GENDER:N', 
            title='Gender', 
            sort='-y', 
            axis=alt.Axis(labelAngle=0)),
    # CHANGE: Map Y directly to the sum of our Total_Count column
    y=alt.Y('sum(Total_Count):Q', title='Number of People'),
    color=alt.Color('GENDER:N', 
                    scale=alt.Scale(range=["#669AE2", "#BC6888"]),
                    legend=None),
    tooltip=['GENDER:N', alt.Tooltip('sum(Total_Count):Q', title='Total Count')]
).transform_filter(
    issue_select
).transform_filter(
    click
).properties(
    width=300,
    height=300,
    title='Gender of Selected Response to Issue'
)

#ARRANGE AND DISPLAY
interactive_chart = alt.hconcat(
    pie, 
    bar, 
    center=True
).resolve_scale(
    color='independent'
).configure_axis(
    labelFontSize=12, # Makes the axis labels bigger
    titleFontSize=14  # Makes the axis titles bigger
).configure_legend(
    labelFontSize=14, # Makes the legend text bigger
    titleFontSize=16  # Makes the legend title bigger
).configure_title(
    fontSize=18       # Makes the main chart titles bigger
)
interactive_chart
```

```python
# Save the chart as a standalone webpage
interactive_chart.save('../presentation/charts/issues_dashboard.html')
```
